## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

Primary Keys are keys of the particular table, and as keys must have a unique value for all entries in the table.
Foreign Keys are primary keys of a different table and can be used to Join two tables together.



## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

One-to-One: a single entry in one table matches up to a single entry in another table. Joined tables will have the same length as both of the original tables.

One-to-Many: a single entry in one table relates to several entries in the second table. The resulting Join would be as long as the larger table.

Many-to-Many: several entries in the first table match to several entries in the second table. The resulting Join would be much larger than the original tables.


## Question 4

Explain the various types of JOINs possible with SQL.

LeftJoin - The first table is the primary and the second is matched up to it
RightJoin - The second table is the primary and the first is matched up to it
InnerJoin - The Join table only contains entries that exist in both tables
OuterJoin - The Join table only contains entries that do not exist in both tables
